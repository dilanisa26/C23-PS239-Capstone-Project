```python
import csv
import numpy as np
import os
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('capstone.csv')
df.head()
```

```python
df[df['class']=='not enough depth']
```

```python
X = df.drop('class', axis=1) # features
y = df['class'] # target value
```

```python
# Splitting the data into training (70%), development (15%), and test (15%) sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
X_dev, X_test, y_dev, y_test = train_test_split(X_test, y_test, test_size=0.5, random_state=42)

# Print the shape of each resulting set
print("Training set shapes:")
print(X_train.shape, y_train.shape)
print("Development set shapes:")
print(X_dev.shape, y_dev.shape)
print("Test set shapes:")
print(X_test.shape, y_test.shape)
```

# Train Machine Learning Classification Model

```python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler

from sklearn.ensemble import RandomForestClassifier
```

```python
pipelines = {
    'rf': make_pipeline(StandardScaler(), RandomForestClassifier())
}
```

```python
fit_models = {}
for algo, pipeline in pipelines.items():
    model = pipeline.fit(X_train,y_train)
    fit_models[algo] = model
```

```python
fit_models
```

```python
fit_models['rf'].predict(X_dev)
```

# Model Evaluation

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import pickle
```

```python
for algo, model in fit_models.items():
    yhat = model.predict(X_test)
    print(algo, classification_report(y_test, yhat))
```
